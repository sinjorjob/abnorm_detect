```python
width = 100   #スライス幅
nk = 1   #クラスタ数
```

```python
#dataをsize毎のスライス窓に分割
def embed(lst, dim):
    emb = np.empty((0,dim), float)
    for i in range(lst.size - dim + 1):
        tmp = np.array(lst[i:i+dim])[::-1].reshape((1,-1)) 
        emb = np.append( emb, tmp, axis=0)
    return emb
```

```python
# サイズ100のスライス窓にデータを分割
train = embed(train_data, width)
test = embed(test_data, width)
```

```python
print(train.shape)
print(test.shape)
```

```python
# 学習データでモデルを生成し、学習データについて異常度（最近傍までの距離を）算出
neigh = NearestNeighbors(n_neighbors=nk)
neigh.fit(train)
# kneighbors(X)メソッドでXに対する近傍の点と距離を得る
# distancesの値が大きい＝距離が離れている＝異常度が高いと判定する
distances, indices = neigh.kneighbors(test)

# 距離をmax1にするデータ整形
# mx = np.max(distances)
# distances = distances / mx
```

```python
# テストデータと異常度のプロット
import matplotlib.pyplot as plt
import japanize_matplotlib
test_for_plot = data[3001+width:6000, 2]
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
p1, = ax1.plot(distances, '-b')
p2, = ax2.plot(test_for_plot, '-g')
ax2.set_ylim(0, 12.0)
plt.title("K近傍法による時系列データ異常度分析")
ax1.legend([p1, p2], ["異常度", "生データ"])
ax1.set_xlabel('時間軸')
ax1.set_ylabel('心電図のデータ')
ax2.set_ylabel('異常度')
plt.grid()
plt.show()
```
