```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
from sklearn.neighbors import NearestNeighbors
```

```python
#データのロード
data = np.loadtxt(r'data\perf.txt',delimiter="\t")
```

```python
print(data.shape)
print(data[0:5])
#メモリ、CPU,CPUキュー値
```

```python
#最初の3000件を訓練データ、次の3000件をテストデータとして抽出
train_data = data[1:3000, 0]
test_data = data[3001:6000, 0]
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
#深さを動的に変化させたときの正解率を描画
import matplotlib.pyplot as plt
import japanize_matplotlib

plt.figure(figsize=(12,3))
x_range = []
for k in range(1, len(train_data)+1):
    x_range.append(k)

plt.plot(x_range, train_data)
plt.grid()
plt.xlabel('時間軸')
plt.ylabel('メモリ使用率（％）')
plt.title('訓練データのグラフ')
plt.ylim(0,100)
plt.show()

#########################################################
x_range = []

for k in range(1, len(test_data)+1):
    x_range.append(k)
plt.figure(figsize=(12,3))
plt.plot(x_range, test_data)
plt.grid()
plt.xlabel('時間軸')
plt.ylabel('メモリ使用率（％）')
plt.title('テストデータのグラフ')
plt.ylim(0,100)
plt.show()
```

```python
width = 100   #スライス幅
nk = 1   #クラスタ数
```

```python
#dataをsize毎のスライス窓に分割
def embed(lst, dim):
    emb = np.empty((0,dim), float)
    for i in range(lst.size - dim + 1):
        tmp = np.array(lst[i:i+dim])[::-1].reshape((1,-1)) 
        emb = np.append( emb, tmp, axis=0)
    return emb
```

```python
# サイズ100のスライス窓にデータを分割
train = embed(train_data, width)
test = embed(test_data, width)
```

```python
print(train.shape)
print(test.shape)
```

```python
# 学習データでモデルを生成し、学習データについて異常度（最近傍までの距離を）算出
neigh = NearestNeighbors(n_neighbors=nk)
neigh.fit(train)
# kneighbors(X)メソッドでXに対する近傍の点と距離を得る
# distancesの値が大きい＝距離が離れている＝異常度が高いと判定する
distances, indices = neigh.kneighbors(test)

# 距離をmax1にするデータ整形
mx = np.max(distances)
distances = distances / mx
```

```python
print(distances.shape)
print(indices.shape)
print(distances.max())
```

```python
print(distances[0:10])
print(indices[0:10])
```

```python
test_for_plot = data[3001+width:6000, 0]
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

p1, = ax1.plot(distances, '-b')
ax1.set_ylabel('異常度')
ax1.set_ylim(0, 1.2)  #異常度の範囲
p2, = ax2.plot(test_for_plot, '-g')
ax2.set_ylabel('メモリ使用率（％）')
ax2.set_ylim(0, 100.0)  #メモリ使用率の範囲
plt.title("k近傍法による異常部位検出")
ax1.legend([p1, p2], ["異常度", "メモリ使用率（％）"])
plt.grid()
plt.show()
```
